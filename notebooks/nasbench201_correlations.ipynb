## Table 1: Spearman ρ of zero-cost proxies on NAS-Bench-201.

```python
t=None
all_ds = {}
all_acc = {}
allc = {}
allk = {}
all_metrics = {}
all_runs = {}
zc_metric_names = [
    'grad_norm', 'snip', 'grasp', 'fisher', 'synflow', 'jacob_cov', 
    'lipschitz', 'lipschitz_supremum', 'synflow_lipschitz']
metric_names = zc_metric_names + ['frobenius', 'nuclear']
for zc in zc_metric_names:
    metric_names.append(f"{zc}_frobenius")
    metric_names.append(f"{zc}_nuclear")

#46
for fname,rname in [('../results_release/nasbench2/nb2_cf100_seed20_dlrandom_dlinfo1_initwnone_initbnone.p','CIFAR10')]:
    runs=[]
    f = open(fname,'rb')
    while(1):
        try:
            runs.append(pickle.load(f))
        except EOFError:
            break
    f.close()
    print(fname, len(runs))
    
    all_runs[fname]=runs
    all_ds[fname] = {}
    metrics={}
    for k in metric_names:
        metrics[k] = []
    acc = []
    
    if t is None:
        hl=['Correlation']
        hl.extend(metric_names)
        t = PrettyTable(hl)
    
    for r in runs:
        for k,v in r['logmeasures'].items():
            if k in metrics:
                metrics[k].append(v)
        acc.append(r['testacc'])
     
    metrics['synflow_lipschitz'] = [x * y for x, y in zip(metrics["synflow"], metrics["lipschitz_supremum"])]
    for zc in zc_metric_names:
        metrics[f'{zc}_frobenius'] = [x * y for x, y in zip(metrics[zc], metrics["frobenius"])]
        metrics[f'{zc}_nuclear'] = [x * y for x, y in zip(metrics[zc], metrics["nuclear"])]
    
    all_ds[fname]['metrics'] = metrics
    all_ds[fname]['acc'] = acc
        
    res = []
    kendall = []
    crs = {}
    kendalls = {}
    for k in hl:
        if k=='Correlation':
            continue
        v = metrics[k]
        new_acc = []
        new_v = []
        remove_outlier = False
        
        if remove_outlier:
            for x in range(len(acc)):
                q75,q25 = np.nanpercentile(v,[75,25])
                intr_qr = q75-q25
                max = q75+(1.5*intr_qr)
                min = q25-(1.5*intr_qr)
                #if (v[x] > min and v[x] < max):
                if acc[x] >= 90:
                    new_v.append(v[x])
                    new_acc.append(acc[x])
        else:
            new_acc = acc
            new_v = v
        cr = stats.spearmanr(new_acc,new_v,nan_policy='omit').correlation
        tau, p_value = stats.kendalltau(new_acc, new_v, nan_policy='omit')
        res.append(round(cr,3))
        kendall.append(round(tau,3))
        crs[k]=cr
        kendalls[k]=tau

    print(rname)
    ds = rname
    all_acc[ds]=acc
    allc[ds]=crs
    allk[ds]=kendalls
    t.add_row(["Spearman"]+res)
    t.add_row(["Kendall tau"]+kendall)
    all_metrics[ds] = metrics
print(t)
```

### Voting between 3 metrics could improve rank correlation

```python
from tqdm import tqdm
votes = {}
def vote(mets, gt):
    numpos = 0
    for m in mets:
        numpos += 1 if m > 0 else 0
    if numpos >= len(mets)/2:
        sign = +1
    else:
        sign = -1
    return sign*gt

for ds in all_acc.keys():
    #num_pts = 15625
    num_pts = 100
    tot=0
    right=0
    for i in tqdm(range(num_pts)):
        for j in range(num_pts):
            if i!=j:
                diff = all_acc[ds][i] - all_acc[ds][j]
                if diff == 0:
                    continue
                diffsyn = []
                for m in ['synflow', 'lipschitz', 'nuclear']:
                    diffsyn.append(all_metrics[ds][m][i] - all_metrics[ds][m][j])
                same_sign = vote(diffsyn, diff)
                right += 1 if same_sign > 0 else 0
                tot += 1
    votes[ds.lower() if 'CIFAR' in ds else ds] = right/tot
print('votes correlation: ', votes)
```

```python
from nas_201_api import NASBench201API as API
api = API('../data/NAS-Bench-201-v1_0-e61699.pth')
api.verbose = False
```

```python
dallb={}
dallb_10={}
dallb_10f={}
for ds in ['cifar10', 'cifar100','ImageNet16-120']:
    allb = {}
    allb_10 = {}
    allb_10f = {}

    for k in range(0,41):

        b=[]
        b_10 = []
        b_10f = []

        for i in tqdm(range(len(api))):
            info = api.get_more_info(i, 'cifar10-valid' if ds=='cifar10' else ds, iepoch=None, hp='200', is_random=False)
            info_10 = api.get_more_info(i, 'cifar10-valid' if ds=='cifar10' else ds, iepoch=k, hp='200', is_random=False)

            try:
                info_10_fast = api.get_more_info(i, 'cifar10-valid' if ds=='cifar10' else ds, iepoch=k, hp='12', is_random=False)
                testacc_10_fast  = info_10_fast['valid-accuracy' if ds=='cifar10' else 'valtest-accuracy']
            except Exception:
                pass
            
            testacc  = info['test-accuracy']
            testacc_10  = info_10['valid-accuracy' if ds=='cifar10' else 'valtest-accuracy']

            b.append(testacc)
            b_10.append(testacc_10)  
            b_10f.append(testacc_10_fast)

        allb[k] = b
        allb_10[k] = b_10
        allb_10f[k] = b_10f
    dallb[ds]=allb
    dallb_10[ds]=allb_10
    dallb_10f[ds]=allb_10f
```

```python
dslow = {}
dfast = {}
for ds,allb in dallb.items():
    dslow[ds] = []
    dfast[ds] = []
    t = PrettyTable(['Epoch', 'Normal Training (200 epochs)', 'Fast Training (12 Epochs)'])
    for k,b in allb.items():
        r = [k]
        for v in [dallb_10[ds][k], dallb_10f[ds][k]]:
            cr = abs(stats.spearmanr(b,v,nan_policy='omit').correlation)
            r.append(round(cr,3))
        t.add_row(r)
        dslow[ds].append(r[1])
        dfast[ds].append(r[2])
```
